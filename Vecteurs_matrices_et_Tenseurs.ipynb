VERY IMPORTANT => Avantage de mettre les data en tableau numpy => la vectorisation qui nous permet d'appliquer à tous les éléments d'un tableau numpy la même transformation !!!

```python
def double(x):        # on implémente une fonction qui marche sur 1 petit élément mais on peut l'exploiter sur tout notre tab. numpy
  return 2 * x
```

```python
double(5)
```

Si on applique notre fonction à une liste d'éléments:

```python
[5, 4, 2]   # et on veut obtenir [10, 8, 4]
```
